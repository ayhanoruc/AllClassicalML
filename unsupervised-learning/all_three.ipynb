```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch # for dendrogram
from sklearn.cluster import AgglomerativeClustering # for hierarchical clustering, we could use devisive approach as well.

# Create a synthetic dataset with 3 clusters
data, true_labels = make_blobs(n_samples=1000, centers=12, cluster_std=1.0, random_state=42)

model = AgglomerativeClustering(n_clusters=6, affinity='euclidean', linkage='ward')
predicted_labels = model.fit_predict(data)

# Visualize the hierarchical clustering results
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, cmap='viridis', alpha=0.5)
plt.title('Hierarchical Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Dendrogram
dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_swiss_roll
from sklearn.cluster import DBSCAN

# Generate a synthetic spiral dataset
data, true_labels = make_swiss_roll(n_samples=1000, noise=0.5, random_state=42)

# Visualize the synthetic spiral dataset
plt.scatter(data[:, 0], data[:, 1], c=true_labels, cmap='viridis', alpha=0.7)
plt.title('Synthetic Spiral Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

```

```python
# Perform DBSCAN clustering
model = DBSCAN(eps=0.2, min_samples=15)
predicted_labels = model.fit_predict(data)

# Visualize the DBSCAN clustering results
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, cmap='viridis', alpha=0.7)
plt.title('DBSCAN Clustering Results on Spiral Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
